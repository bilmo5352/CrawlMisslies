```python
import requests
import time

URL = "https://product-extraction-pipeline-production.up.railway.app/health"

def ping():
    try:
        response = requests.get(URL, timeout=10)
        print(f"[PING] Status: {response.status_code} | {response.text}")
    except Exception as e:
        print(f"[ERROR] {e}")

def main():
    while True:
        ping()
        # ping every 10 minutes (600 seconds)
        time.sleep(6)

if __name__ == "__main__":
    main()
```
