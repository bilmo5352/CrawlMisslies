```python
import wikipedia

# search for a category or keyword
results = wikipedia.search("smart phone and appliances")
print(results)  # shows page titles related to electronics

# get content of a specific page
page = wikipedia.page("List of consumer electronics")
print(page.title)
print(page.content[:500])  # preview

# get links on that page (which often are product names)
print(page.links[:20])
```

```python
import wikipediaapi

wiki = wikipediaapi.Wikipedia("en")

# Access a category
category = wiki.page("Category:Consumer electronics")

def print_categorymembers(categorymembers, level=0, max_level=1):
    for c in categorymembers.values():
        print("%s: %s" % ("  " * level, c.title))
        if c.ns == wikipediaapi.Namespace.CATEGORY and level < max_level:
            print_categorymembers(c.categorymembers, level + 1)

print_categorymembers(category.categorymembers)
```

```python
import wikipediaapi

wiki = wikipediaapi.Wikipedia('en')
category = wiki.page("Category:Smartphones")

def get_product_names(cat_page):
    products = []
    for title, page in cat_page.categorymembers.items():
        if page.ns == wikipediaapi.Namespace.MAIN:
            products.append(title)
    return products

products = get_product_names(category)
print(products[:30])
```
